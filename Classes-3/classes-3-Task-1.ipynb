```python
def Hamiltonian_matrix(size):
    """
    working only in GOE casses (see lecture)
    """
    h = np.random.randn(size,size)
    return (h+h.T)/2

def Find_eigenValues(matrix):
    eigenVal, eigenVectors = eig(matrix)
    return eigenVal
```

```python
N = 6

h = Hamiltonian_matrix(N)
print(h)
print()

eigVals = Find_eigenValues(h)
print(eigVals)
```

# Set inital conditions: get $n_{\mathrm{samples}} = 20 000$

```python
def Wigner(_E, _R):
    return 2 / (np.pi * _R ** 2) * np.sqrt(_R**2 - _E**2)
```

```python
N = 6 # N = 20, 200
nSamps = 20000 # nSamps = 10000, 500
R = np.sqrt(2 * N)

whole_eigenVals = np.array([])

for i in range(0, nSamps):
    # randomise hamiltonian matrices
    h = Hamiltonian_matrix(N)
    # find eigenvalues
    eigenVals = Find_eigenValues(h)
    # appending to list cotains eigenvalues
    whole_eigenVals = np.append(whole_eigenVals, eigenVals)
```

```python
# print(whole_eigenVals)
```

```python
"""
Plot histogram
"""

fig, ax = plt.subplots(figsize=(9.0, 6.0))
# data
n, bins, cos = plt.hist(whole_eigenVals, 50, density=True, facecolor='cyan', alpha=0.75)
plt.plot(bins, Wigner(bins, R), 'r-', linewidth=2)
# describing
plt.title(r'Frequency of occurrence Eigenvalue', fontsize=20)
plt.ylabel('Normalised occurrence', fontsize=20)
plt.xlabel(r'Eigenvalue: $E$', fontsize=20)
plt.show()
```
